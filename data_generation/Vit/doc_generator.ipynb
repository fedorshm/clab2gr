```python
# Основной генератор
import os
import random
from collections import OrderedDict
from docx import Document
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.enum.section import WD_ORIENT
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
from docx.shared import Pt, RGBColor, Inches, Cm
from faker import Faker
from mimesis import Datetime
import sympy as sp
from sympy import symbols, Matrix
import matplotlib.pyplot as plt
from PIL import Image

# Инициализация генераторов
locales = OrderedDict([('en-US', 1), ('ru-RU', 2)])
fake = Faker(locales)
datetime_gen = Datetime()

COLORS = """000000 000080 00008B 0000CD 0000FF 006400 008000 008080 008B8B 00BFFF 00CED1 
00FA9A 00FF00 00FF7F 00FFFF 00FFFF 191970 1E90FF 20B2AA 228B22 2E8B57 2F4F4F 32CD32 3CB371 
40E0D0 4169E1 4682B4 483D8B 48D1CC 4B0082 556B2F 5F9EA0 6495ED 66CDAA 696969 6A5ACD 6B8E23 
708090 778899 7B68EE 7CFC00 7FFF00 7FFFD4 800000 800080 808000 808080 87CEEB 87CEFA 8A2BE2 
8B0000 8B008B 8B4513 8FBC8F 90EE90 9370D8 9400D3 98FB98 9932CC 9ACD32 A0522D A52A2A A9A9A9 
ADD8E6 ADFF2F AFEEEE B0C4DE B0E0E6 B22222 B8860B BA55D3 BC8F8F BDB76B C0C0C0 C71585 CD5C5C 
CD853F D2691E D2B48C D3D3D3 D87093 D8BFD8 DA70D6 DAA520 DC143C DCDCDC DDA0DD DEB887 E0FFFF 
E6E6FA E9967A EE82EE EEE8AA F08080 F0E68C F0F8FF F0FFF0 F0FFFF F4A460 F5DEB3 F5F5DC F5F5F5 
F5FFFA F8F8FF FA8072 FAEBD7 FAF0E6 FAFAD2 FDF5E6 FF0000 FF00FF FF00FF FF1493 FF4500 FF6347 
FF69B4 FF7F50 FF8C00 FFA07A FFA500 FFB6C1 FFC0CB FFD700 FFDAB9 FFDEAD FFE4B5 FFE4C4 FFE4E1 
FFEBCD FFEFD5 FFF0F5 FFF5EE FFF8DC FFFACD FFFAF0 FFFAFA FFFF00 FFFFE0 FFFFF0 FFFFFF"""

colors_list = COLORS.split()

# Папка для изображений
images_folder = "C:\\Users\\Vito\\Jupyter\\Генератор\\Рисунки"  # путь к папке с изображениями

# Переменные для уравнений
x, y, z, a, b, c = sp.symbols('x y z a b c')

def generate_random_formula():
    # Типы операций и функции
    operations = ['+', '-', '*', '/', '**', 'mod']
    functions = ['sqrt', 'log', 'exp', 'sin', 'cos', 'tan', 'Abs', 'factorial']
    
    # Виды формул
    formula_type = random.choice(['basic', 'function', 'equation', 'calculus', 'fraction', 'complex'])
    
    if formula_type == 'basic':
        num1, num2 = random.randint(1, 10), random.randint(1, 10)
        operation = random.choice(operations)
        formula = f"{num1} {operation} {num2}"
        result = eval(formula.replace("mod", "%"))  # Модульная операция
        return f"{formula} = {result}"
    
    elif formula_type == 'function':
        func = random.choice(functions)
        num = random.randint(1, 10)
        if func == 'sqrt':
            formula = f"√({num})"
            result = sp.sqrt(num)
        elif func == 'log':
            base = random.choice([2, 10, sp.E])
            formula = f"log_{base}({num})"
            result = sp.log(num, base)
        elif func == 'exp':
            formula = f"exp({num})"
            result = sp.exp(num)
        elif func == 'factorial':
            formula = f"{num}!"
            result = sp.factorial(num)
        else:
            formula = f"{func}({num})"
            result = getattr(sp, func)(num)
        return f"{formula} = {result}"
    
    elif formula_type == 'equation':
        coeff1, coeff2, constant = random.randint(1, 5), random.randint(1, 5), random.randint(1, 10)
        var = random.choice([x, y])
        if random.choice(['linear', 'quadratic']) == 'linear':
            formula = f"{coeff1}*{var} + {coeff2} = {constant}"
        else:
            formula = f"{coeff1}*{var}**2 + {coeff2}*{var} + {constant} = 0"
        return formula
    
    elif formula_type == 'calculus':
        operation = random.choice(['derivative', 'integral'])
        func = random.choice(['x**2', 'sin(x)', 'e**x', 'x*log(x)'])
        variable = x
        if operation == 'derivative':
            formula = f"d/dx({func})"
            result = sp.diff(func, variable)
        else:
            formula = f"∫({func})dx"
            result = sp.integrate(func, variable)
        return f"{formula} = {result}"
    
    elif formula_type == 'fraction':
        num, denom = random.randint(1, 10), random.randint(1, 10)
        formula = sp.Rational(num, denom)
        return f"{formula} = {num / denom:.2f}"
    
    elif formula_type == 'complex':
        formula = random.choice([
            f"{x}**2 + {y}**2",
            f"sin({x}) + cos({y})",
            f"log({x} + {y})",
            f"{x}*{y} + {y}*{z}"
        ])
        return f"Complex expression: {formula}"
    
    return "No formula generated"

def create_formula_image(formula, image_path, width=2, height=1):
    """
    Создаёт изображение формулы фиксированного размера.
    """
    dpi = 300  # Разрешение изображения
    plt.figure(figsize=(width, height), dpi=dpi)
    plt.text(0.5, 0.5, f"${formula}$", fontsize=30, ha='center', va='center')
    plt.axis('off')
    plt.savefig(image_path, bbox_inches='tight', pad_inches=0, dpi=dpi)
    plt.close()
    
    # Пропорционально изменяем размер изображения (в пикселях)
    target_width = int(width * dpi)
    target_height = int(height * dpi)
    with Image.open(image_path) as img:
        img = img.resize((target_width, target_height), Image.Resampling.LANCZOS)  # Используем LANCZOS вместо ANTIALIAS
        img.save(image_path)

def add_formula_image_to_document(document):
    """
    Вставляет изображение формулы фиксированного размера в документ.
    """
    formula = generate_random_formula()  # Генерация случайной формулы
    image_path = os.path.join(images_folder, "formula.png")
    
    # Создание изображения формулы с фиксированными размерами
    create_formula_image(formula, image_path, width=2, height=1)
    
    # Вставка изображения в документ с фиксированной шириной
    paragraph = document.add_paragraph()
    run = paragraph.add_run()
    run.add_picture(image_path, width=Inches(2))  # Устанавливаем ширину изображения
    
    # Выравнивание изображения по центру
    paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

# Функция для смены ориентации страницы
def change_orientation(document):
    current_section = document.sections[-1]
    new_width, new_height = current_section.page_height, current_section.page_width
    current_section.orientation = WD_ORIENT.LANDSCAPE
    current_section.page_width = new_width
    current_section.page_height = new_height

def set_margins(section):
    """
    Устанавливает отступы слева и справа на 0.5 см для секции.
    """
    section.left_margin = Cm(1.25)  # Отступ слева 0.5 см
    section.right_margin = Cm(1.25)  # Отступ справа 0.5 см

def add_spacing(document, lines=1):
    """
    Добавляет отступ в документ.
    
    :param document: объект документа.
    :param lines: количество пустых строк (по умолчанию 1).
    """
    for _ in range(lines):
        paragraph = document.add_paragraph()  # Создаём пустой параграф
        paragraph.paragraph_format.space_after = Pt(0)  # Убираем лишние отступы после параграфа

# Добавление колонтитулов с одинаковым выравниванием
def add_headers_and_footers(document):
    # Случайное выравнивание для колонтитулов
    alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT, 
                               WD_PARAGRAPH_ALIGNMENT.CENTER, 
                               WD_PARAGRAPH_ALIGNMENT.RIGHT])

    # Настройка верхнего колонтитула
    header = document.sections[0].header.paragraphs[0]
    header.text = fake.sentence(nb_words=5)
    header.alignment = alignment  # Применяем случайное выравнивание

    # Настройка нижнего колонтитула
    footer = document.sections[0].footer.paragraphs[0]
    footer.text = fake.sentence(nb_words=5)
    footer.alignment = alignment  # Применяем то же выравнивание

# Добавление заголовка
def add_title(document, title_font_size):
    title = document.add_paragraph()
    include_numbering = random.choice([True, False])
    number_prefix = f"{random.randint(1, 10)}. " if include_numbering else ""
    run = title.add_run(number_prefix + fake.sentence(nb_words=5))
    run.bold = True
    run.italic = random.choice([True, False])
    run.font.size = Pt(title_font_size)
    title.alignment = random.choices([WD_PARAGRAPH_ALIGNMENT.CENTER, 
                                       WD_PARAGRAPH_ALIGNMENT.LEFT, 
                                       WD_PARAGRAPH_ALIGNMENT.RIGHT], 
                                      weights=[0.7, 0.15, 0.15])[0]
    document.add_paragraph()

# Добавление текста с красной строкой и случайным выравниванием
def add_paragraph(document, base_font_size):
    paragraph = document.add_paragraph()
    run = paragraph.add_run(fake.text(max_nb_chars=random.randint(200, 800)))
    run.font.size = Pt(base_font_size)
    alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT, 
                               WD_PARAGRAPH_ALIGNMENT.CENTER, 
                               WD_PARAGRAPH_ALIGNMENT.RIGHT, 
                               WD_PARAGRAPH_ALIGNMENT.JUSTIFY])
    paragraph.alignment = alignment
    paragraph.paragraph_format.first_line_indent = Pt(12)

def get_table_styles(document):
    """Возвращает список доступных стилей таблиц в документе, исключая стили других типов."""
    styles = document.styles
    table_styles = [style.name for style in styles if style.type == 3]  # Проверяем, что стиль относится к таблицам
    return table_styles

def add_table_with_caption(document, base_font_size, caption_font_size):
    # Получаем список стилей таблиц
    table_styles = get_table_styles(document)

    # Создаем подпись таблицы с случайным выравниванием
    caption = document.add_paragraph(f"Табл. {random.randint(1, 100)} - {fake.sentence(nb_words=4)}")
    caption.alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT,
                                       WD_PARAGRAPH_ALIGNMENT.CENTER,
                                       WD_PARAGRAPH_ALIGNMENT.RIGHT])
    caption.runs[0].font.size = Pt(caption_font_size)  # Устанавливаем размер шрифта для подписи

    # Создаем таблицу с случайным числом строк и столбцов
    table = document.add_table(rows=random.randint(5, 10), cols=random.randint(4, 8))

    # Случайный выбор стиля таблицы
    if table_styles:
        table.style = random.choice(table_styles)

    # Случайное выравнивание для самой таблицы
    table.alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT,
                                     WD_PARAGRAPH_ALIGNMENT.CENTER,
                                     WD_PARAGRAPH_ALIGNMENT.RIGHT])

    # Выбор черно-белого варианта или цветного
    if random.choice([True, False]):  # 0.5 вероятность черно-белой таблицы
        color = "FFFFFF"  # Белый цвет для фона
        font_color = RGBColor(0, 0, 0)  # Черный цвет для текста
    else:
        # Выбор цвета для всей таблицы
        color = random.choice(colors_list)
        font_color = RGBColor(255, 255, 255) if color in ["000000", "000080", "800000"] else RGBColor(0, 0, 0)

    # Заполняем ячейки таблицы случайными данными
    for row in table.rows:
        for cell in row.cells:
            # Случайный выбор данных для заполнения ячейки
            data_type = random.choice(['text', 'number', 'date'])
            if data_type == 'text':
                cell.text = fake.text(max_nb_chars=random.randint(10, 50))
            elif data_type == 'number':
                cell.text = str(fake.random_number())
            else:  # data_type == 'date'
                cell.text = str(fake.date())

            # Устанавливаем размер шрифта для текста в ячейке
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.size = Pt(base_font_size)
                    run.font.color.rgb = font_color  # Применяем цвет текста

                # Применяем случайное выравнивание к параграфу внутри ячейки
                paragraph.alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT,
                                                     WD_PARAGRAPH_ALIGNMENT.CENTER,
                                                     WD_PARAGRAPH_ALIGNMENT.RIGHT])

            # Заливка ячейки цветом
            shading = OxmlElement('w:shd')
            shading.set(qn('w:fill'), color)  # Устанавливаем цвет фона
            cell._element.get_or_add_tcPr().append(shading)

    # Убираем левые и/или правые границы для всей первой и последней колонок, если таблица имеет сетку
    if random.choice([True, False]):  # 0.5 вероятность применения изменений к границам
        remove_left_border = random.choice([True, False])  # Случайно убрать левую границу
        remove_right_border = random.choice([True, False])  # Случайно убрать правую границу

        for row in table.rows:
            if remove_left_border:
                cell = row.cells[0]
                tcPr = cell._element.get_or_add_tcPr()
                borders = OxmlElement('w:tcBorders')
                left_border = OxmlElement('w:left')
                left_border.set(qn('w:val'), 'nil')  # Убираем левую границу
                borders.append(left_border)
                tcPr.append(borders)

            if remove_right_border:
                cell = row.cells[-1]
                tcPr = cell._element.get_or_add_tcPr()
                borders = OxmlElement('w:tcBorders')
                right_border = OxmlElement('w:right')
                right_border.set(qn('w:val'), 'nil')  # Убираем правую границу
                borders.append(right_border)
                tcPr.append(borders)

# Добавление случайного текста
def add_random_text(document, base_font_size):
    paragraph = document.add_paragraph()
    run = paragraph.add_run(fake.text(max_nb_chars=random.randint(200, 800)))
    run.font.size = Pt(base_font_size)
    alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT, 
                               WD_PARAGRAPH_ALIGNMENT.CENTER, 
                               WD_PARAGRAPH_ALIGNMENT.RIGHT, 
                               WD_PARAGRAPH_ALIGNMENT.JUSTIFY])
    paragraph.alignment = alignment
    paragraph.paragraph_format.first_line_indent = Pt(12)

def add_two_column_text(document, base_font_size):
    """
    Добавляет текст в две колонки в текущем разделе документа.
    """
    # Настройка двух колонок через XML
    current_section = document.sections[-1]
    sectPr = current_section._sectPr
    cols = sectPr.xpath('./w:cols')[0]
    cols.set(qn('w:num'), '2')  # Устанавливаем две колонки

    # Генерация текста для вставки
    for _ in range(random.randint(3, 6)):  # Количество абзацев
        paragraph = document.add_paragraph(fake.text(max_nb_chars=random.randint(500, 1000)))
        paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.JUSTIFY  # Выравнивание текста по ширине
        paragraph.paragraph_format.space_after = Pt(4)  # Отступ после абзаца

        # Применяем размер шрифта для каждого абзаца
        for run in paragraph.runs:
            run.font.size = Pt(base_font_size)

# Добавление изображения с подписью
def add_image_with_caption(document, caption_font_size):
    image_path = os.path.join(images_folder, random.choice(os.listdir(images_folder)))
    
    # Ограничиваем ширину изображения до 40% ширины страницы, чтобы оно не выходило за рамки
    section_width = document.sections[0].page_width
    max_image_width = section_width * 0.4  # 40% от ширины страницы
    
    # Вставляем изображение с ограничением по ширине
    document.add_picture(image_path, width=max_image_width)

    # Добавление подписи
    caption = document.add_paragraph(f"Рис. {random.randint(1, 100)} - {fake.sentence(nb_words=6)}")
    caption.alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT, 
                                      WD_PARAGRAPH_ALIGNMENT.CENTER, 
                                      WD_PARAGRAPH_ALIGNMENT.RIGHT])
    caption.runs[0].font.size = Pt(caption_font_size)


# Добавление нумерованного списка с одинаковым выравниванием
def add_numbered_list(document, base_font_size):
    list_alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT,
                                     WD_PARAGRAPH_ALIGNMENT.CENTER,
                                     WD_PARAGRAPH_ALIGNMENT.RIGHT])  # Случайное выравнивание для всего списка
    for _ in range(random.randint(3, 7)):
        paragraph = document.add_paragraph(fake.sentence(nb_words=6), style='List Number')
        paragraph.alignment = list_alignment

        # Применяем размер шрифта для каждого элемента списка
        for run in paragraph.runs:
            run.font.size = Pt(base_font_size)

# Добавление маркированного списка
def add_bulleted_list(document, base_font_size):
    # Случайное выравнивание для всего списка
    list_alignment = random.choice([WD_PARAGRAPH_ALIGNMENT.LEFT,
                                     WD_PARAGRAPH_ALIGNMENT.CENTER,
                                     WD_PARAGRAPH_ALIGNMENT.RIGHT])  # случайное выравнивание
    for _ in range(random.randint(3, 7)):
        paragraph = document.add_paragraph(fake.sentence(nb_words=6), style='List Bullet')
        paragraph.alignment = list_alignment  # Применяем одинаковое выравнивание ко всем элементам списка

        # Применяем размер шрифта для каждого элемента списка
        for run in paragraph.runs:
            run.font.size = Pt(base_font_size)

# Функция для создания документа
def create_document():
    document = Document()

    # Устанавливаем отступы для первой секции
    set_margins(document.sections[0])

    # Генерация размеров шрифтов
    base_font_size = random.randint(8, 16)  # Размер основного шрифта
    title_font_size = base_font_size + random.randint(1, 4)  # Заголовок больше основного текста
    caption_font_size = base_font_size - random.randint(1, 2)  # Подписи меньше основного текста
    
    add_headers_and_footers(document)
    if random.choice([True, False]):
        change_orientation(document)
    add_title(document, title_font_size)  # Передаём размер шрифта для заголовка
    add_spacing(document)  # Отступ
    add_paragraph(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ
    add_numbered_list(document, base_font_size)  # Список с основным шрифтом
    add_spacing(document)  # Отступ
    add_random_text(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ
    add_table_with_caption(document, base_font_size, caption_font_size)  # Таблица с подписями
    add_spacing(document)  # Отступ
    add_random_text(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ
    add_random_text(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ
    add_image_with_caption(document, caption_font_size)  # Изображение с подписью
    add_spacing(document)  # Отступ
    add_random_text(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ

    # Добавляем двухколоночный текст с вероятностью 50%
    if random.choice([True, False]):
        add_two_column_text(document, base_font_size)  # Двухколоночный текст

    add_random_text(document, base_font_size)  # Основной текст
    add_spacing(document)  # Отступ
    add_bulleted_list(document, base_font_size)  # Маркированный список
    add_spacing(document)  # Отступ
    add_formula_image_to_document(document)  # Добавляем случайную формулу в документ

    # Устанавливаем отступы для всех секций (на случай, если в документе несколько секций)
    for section in document.sections:
        set_margins(section)
    
    return document

# Генерация документов
output_folder = "C:\\Users\\Vito\\Jupyter\\Генератор\\Сгенерированные документы"
os.makedirs(output_folder, exist_ok=True)
for i in range(40):
    doc = create_document()
    doc.save(os.path.join(output_folder, f"generated_doc_{i+1}.docx"))
```

```python
# вставка сносок в документы
import win32com.client as win32
import random
import os
from faker import Faker
import time

# Инициализация Faker для генерации случайного текста
fake = Faker()

# Функция для добавления сносок
def add_footnotes(word_doc, num_footnotes=3):
    """
    Добавляет сноски в случайные места в документе Word.
    """
    paragraphs = word_doc.Paragraphs
    num_paragraphs = paragraphs.Count
    footnotes_added = 0  # Счетчик добавленных сносок

    # Цикл до тех пор, пока не добавим нужное количество сносок
    while footnotes_added < num_footnotes:
        # Выбор случайного параграфа для добавления сноски
        paragraph_index = random.randint(1, num_paragraphs)
        paragraph = paragraphs(paragraph_index)
        
        # Проверка, что параграф не является частью таблицы
        if not paragraph.Range.Information(win32.constants.wdWithInTable):
            # Добавляем сноску к этому параграфу
            footnote = word_doc.Footnotes.Add(Range=paragraph.Range)
            # Заполняем сноску случайным текстом
            footnote.Range.Text = fake.text(max_nb_chars=100)
            footnotes_added += 1  # Увеличиваем счетчик добавленных сносок
        else:
            # Пропускаем, если параграф внутри таблицы, и ищем следующий подходящий
            continue

# Функция для обработки всех документов в папке
def process_documents_in_folder(folder_path):
    # Инициализация Word через COM
    word_app = win32.gencache.EnsureDispatch('Word.Application')
    word_app.Visible = False  # Запуск Word в фоновом режиме

    # Получаем список всех файлов в папке
    for filename in os.listdir(folder_path):
        if filename.endswith(".docx"):
            file_path = os.path.join(folder_path, filename)
            
            try:
                # Открываем документ
                doc = word_app.Documents.Open(file_path)
                
                # Добавляем сноски
                add_footnotes(doc)

                # Сохраняем документ
                doc.Save()

                print(f"Сноски добавлены в документ: {file_path}")

            except Exception as e:
                print(f"Ошибка при обработке документа {file_path}: {e}")
            
            finally:
                # Закрываем документ
                if 'doc' in locals() and doc:
                    doc.Close()

    # Закрытие Word только после обработки всех документов
    word_app.Quit()

# Указание пути к папке с документами
folder_path = "C:\\Users\\Vito\\Jupyter\\Генератор\\Сгенерированные документы"

# Запуск функции обработки документов
process_documents_in_folder(folder_path)
```
